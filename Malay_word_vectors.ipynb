```python
import logging
import gensim
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
data_file = "data.txt"
def read_input(input_file):
    with open (input_file, 'rb') as f:
        for i, line in enumerate (f): 
            if (i%500000==0):
                logging.info ("read {0} articles".format (i))
            yield gensim.utils.simple_preprocess (line)
# convert articles to list and remove empty lists
documents = [doc for doc in list (read_input (data_file)) if doc != []]
logging.info ("Done")
```

```python
print(documents[2])
```

```python
logger = logging.getLogger()
logger.setLevel(logging.INFO)
```

```python
#Min word count = 100 words

model = gensim.models.Word2Vec (documents, size=100, window=10, min_count=100, workers=10)
model.train(documents,total_examples=len(documents),epochs=10)
```

```python
#save the model, so we can reload it without retraining
model.save("word2vec.model")
```
